```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
from sklearn import linear_model, svm   
from sklearn.metrics import accuracy_score
```

```python
df_train = pd.read_csv('../data/train_data.csv')
df_test = pd.read_csv('../data/test_data.csv')
```

```python
check_point = len(df_train)
```

```python
df = pd.concat([df_train, df_test])
```

```python
df.iloc[check_point]
```

```python
df.isnull().sum()
```

```python
df.replace([r' <=50K', r' <=50K.'], 0, inplace=True)
df.replace([r' >50K', r' >50K.'], 1,inplace=True)
```

```python
df.mode()
```

```python
df['workclass'] = df['workclass'].fillna(df['workclass'].mode()[0])
df['occupation'] = df['occupation'].fillna(df['occupation'].mode()[0])
df['native-country'] = df['native-country'].fillna(df['native-country'].mode()[0])
```

```python
df.info()
```

```python
df = pd.get_dummies(df)
```

```python
df
```

```python
df['salary'].unique()
```

```python
train_data = df.iloc[: check_point]
test_data = df.iloc[check_point :]
```

```python
train_data
```

```python
X_train = train_data.loc[: , ~train_data.columns.isin(['salary'])].values
y_train = train_data['salary'].values
X_test = test_data.loc[: , ~test_data.columns.isin(['salary'])].values
y_test = test_data['salary'].values
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
train_data['salary']
```

```python
model = linear_model.LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
model_2 = svm.SVC(kernel='linear',C = 1)
model_2.fit(X_train, y_train)
```
