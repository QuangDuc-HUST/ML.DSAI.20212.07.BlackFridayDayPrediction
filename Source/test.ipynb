```python
import pandas as pd
import os
from sklearn import linear_model
from sklearn.metrics import accuracy_score
import numpy as np

current = os.getcwd() 
train_file = pd.read_csv(current + r"\..\rawdata\train.csv")
test_file = pd.read_csv(current  + r"\..\rawdata\test.csv")

# file = pd.concat([train_file, test_file])
# file.head()
train_file = train_file.loc[:, ~train_file.columns.isin(['User_ID', 'Product_ID'])]
test_file = test_file.loc[:, ~test_file.columns.isin(['User_ID', 'Product_ID'])]
checkpoint = train_file.shape[0]
# print(checkpoint)
train_file.head()

```

```python
train_file.info()
```

```python
file = pd.concat([train_file, test_file])

category_2_X_data = file.loc[~file['Product_Category_2'].isna(), ~file.columns.isin(['Product_Category_2', 'Product_Category_3', 'Purchase'])]
category_2_Y_data = file.loc[~file['Product_Category_2'].isna(), 'Product_Category_2']
print(category_2_X_data.shape)
# category_2_X_data.head()

```

```python
category_2_X_data = pd.get_dummies(category_2_X_data)
category_2_X_data
```

```python
# category_2_Y_data = pd.get_dummies(category_2_Y_data)
category_2_Y_data
```

```python
model = linear_model.LogisticRegression(max_iter=200)
model.fit(category_2_X_data.values, category_2_Y_data.values)
```

```python
category_2_X_pred = file.loc[file['Product_Category_2'].isna(), ~file.columns.isin(['Product_Category_2', 'Product_Category_3', 'Purchase'])]
category_2_X_pred = pd.get_dummies(category_2_X_pred)
category_2_X_pred
```

```python
category_2_Y_pred = model.predict(category_2_X_pred)
len(category_2_Y_pred)
```

```python
file.loc[file['Product_Category_2'].isna(), 'Product_Category_2'] = category_2_Y_pred
```

```python
file.info()
```

```python
category_3_X_data = file.loc[~file['Product_Category_3'].isna(), ~file.columns.isin(['Product_Category_2', 'Product_Category_3', 'Purchase'])]
category_3_Y_data = file.loc[~file['Product_Category_3'].isna(), 'Product_Category_3']
```

```python
category_3_X_data = pd.get_dummies(category_3_X_data)
category_3_X_data
```

```python
model1 = linear_model.LogisticRegression(max_iter=200)
model1.fit(category_3_X_data, category_3_Y_data)
```

```python
category_3_X_pred = file.loc[file['Product_Category_3'].isna(), ~file.columns.isin(['Product_Category_2', 'Product_Category_3', 'Purchase'])]
category_3_X_pred = pd.get_dummies(category_3_X_pred)
category_3_X_pred
```

```python
category_3_Y_pred = model1.predict(category_3_X_pred)
category_3_Y_pred
```

```python
file.loc[file['Product_Category_3'].isna(), 'Product_Category_3'] = category_3_Y_pred
file.info()
```

```python
file = pd.get_dummies(file)
```

```python
train_data = file[:checkpoint]
test_data = file[checkpoint:]
```

```python
train_X = train_data.loc[:, ~train_data.columns.isin(['Purchase'])].values
train_Y = train_data['Purchase'].values
```

```python
model2 = linear_model.LinearRegression()
model2.fit(train_X, train_Y)
```

```python
test_X = test_data.loc[:, ~train_data.columns.isin(['Purchase'])].values
y_pred = model2.predict(test_X)
y_pred
```
